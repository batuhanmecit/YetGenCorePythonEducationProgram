# Problem 1
Vehicle sınıfından miras alan bir Bus sınıfı oluşturun. Bus.seating_capacity()'nin kapasite bağımsız değişkenine varsayılan olarak 50 değerini verin.

```
Output:
> The seating capacity of a bus is 50 passengers
```


````python
class Vehicle:
    def __init__(self, marka, model, yas):
        self.marka = marka
        self.model = model
        self.yas = yas
        print("Vehicle nesnesi oluşturuldu")

class Bus(Vehicle):
    def __init__(self, marka, model, yas, seating_capacity = 50):
        super().__init__(marka, model, yas)
        self.seating_capacity = seating_capacity
        print("Bus nesnesi oluşturuldu.")
    
    def seating(self):
        print("Seating capacity oluşturuldu")
        print(f"The seating capacity of a bus is {self.seating_capacity} passengers.")
````

````python
b2 = Bus("A" , "B" , 2)
````

````python
b2.seating()
````
